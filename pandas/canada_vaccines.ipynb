```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.0f}'.format
```

```python
canada_shot = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/country_data/Canada.csv')
canada_shot.tail()
```

```python
canada_vac = canada_shot[['date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'location']]
canada_vac.rename(columns={'location' : 'country'}, inplace=True)
canada_vac.tail()
```

```python
canada_vac.to_csv('canada_shot.csv', index=False)
```
