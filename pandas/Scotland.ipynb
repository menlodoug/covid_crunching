```python
import pandas as pd
URL = 'https://www.opendata.nhs.scot/api/3/action/datastore_search?resource_id=e8454cf0-1152-4bcb-b9da-4343f625dfef'
df = pd.read_json(URL)
df
```

```python
s=df['result']
la = s['records']
la
```

```python
scot_df = pd.DataFrame(la)
scot_df
```

```python
scot_df['CA'] = scot_df['CA'].replace(['S12000050','S12000049','S12000048','S12000047'],['S12000044','S12000046','S12000024','S12000015'])
```

```python
scot_df.drop('_id', axis=1, inplace=True)
scot_df
```

```python
population = pd.read_csv('scot_populations.csv')
population['Population'] = population['Population'].round(0).astype(int)
population
```

```python
scotland_df = pd.merge(scot_df, population,  how='right', on='CA')
scotland_df
```

```python
scotland_df.to_csv('scotland_uk.csv', index=False)
```

```python
URL4 = 'https://www.opendata.nhs.scot/api/3/action/datastore_search?resource_id=9393bd66-5012-4f01-9bc5-e7a10accacf4&limit=15000'
df4 = pd.read_json(URL4)
df4
```

```python
s4=df4['result']
sex_age = s4['records']
sex_age
```

```python
new_df.drop(['_id', 'SexQF', 'AgeGroupQF'], axis=1, inplace=True)
new_df
```

```python
new_df.to_csv('scots_by_age_sex.csv', index=False)
```
