```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.0f}'.format
```

```python
country_shot = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv')
country_shot.tail()
```

```python
country_vac = country_shot[['date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'location']]
country_vac.rename(columns={'location' : 'country'}, inplace=True)
country_vac.tail()
```

```python
country_continent = pd.read_csv('country.csv')
country_continent.head()
```

```python
world_df = pd.read_csv('World.csv', encoding='windows-1252')
world_df
```

```python
country_vaccination = pd.merge(country_vac, country_continent, on='country', how='outer')
country_vaccination
```

```python
df =  pd.merge(country_vaccination, world_df, on='country', how ='outer')
df
```

```python
df.to_csv('country_shot.csv', index=False)
```

```python
uk_shot = df.loc[(df['country'] == 'Scotland') | (df['country'] == 'England') | (df['country'] == 'Wales') | (df['country'] == 'Northern Ireland')]
uk_shot.to_csv('uk_shot.csv', index=False)
```
