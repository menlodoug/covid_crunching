```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
```

```python
google_df = pd.read_csv('Global_Mobility_Report.csv', low_memory=False)
google_df['census_fips_code'] = google_df['census_fips_code'].fillna(0).astype(np.int64)
google_df.tail()
```

```python
us_google = google_df[google_df['country_region'] == 'United States']
us_google["sub_region_2"]= us_google["sub_region_2"].str.replace(" County", "")
us_google["sub_region_2"] = us_google["sub_region_2"].fillna('Statewide')
us_google
```
