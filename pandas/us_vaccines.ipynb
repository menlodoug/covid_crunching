```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.0f}'.format
```

```python
us_shot = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/country_data/United%20States.csv')
us_shot.head()
```

```python
us_vac = us_shot[['date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'location']]
us_vac.rename(columns={'location' : 'country'}, inplace=True)
us_vac
```

```python
us_vac.to_csv('us_shot.csv', index=False)
```
