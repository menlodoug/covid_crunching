```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.3f}'.format
```

```python
shots_df = pd.read_csv('covid19_vaccinations.csv')
shots_df
```

```python
shots_df['Pop'] = shots_df['Total Doses Distributed']/(shots_df['Doses Distributed per 100K']/100000)
shots_df.head()
```

```python
booster_df = shots_df[['State/Territory/Federal Entity', 'Total Doses Distributed', 'Doses Distributed per 100K', 'People with a first booster dose','Percent of fully vaccinated people with a first booster dose' ,'Pop', 'Date' ]]
booster_df.head()
```

```python
#booster_df['Booster Percent'] = booster_df['People with a first booster dose']/(booster_df['Pop'])
booster_df.head()
```

```python
booster_df = booster_df.rename(columns = {'State/Territory/Federal Entity' : 'State', 'Total Doses Distributed' : 'Total Doses', 'Doses Distributed per 100K' : 'Doses per 100K', 'People who have received a booster dose' : 'Booster Doses'})
booster_df.head()
```

```python
# Export updated national data to csv for use in Tableau

booster_df.to_csv('booster.csv', index=False)
```
