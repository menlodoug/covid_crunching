```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.2f}'.format
```

```python
sn = pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-nursing-home-county-totals.csv')
sn['state'] = 'California'
sn['state2'] = 'CA'
sn.rename(columns = {'minimum_staff_confirmed_cases':'Staff Confirmed Cases', 'minimum_patients_confirmed_cases':'Patients Confirmed Cases', 
                              'minimum_staff_deaths':'Staff Deaths', 'minimum_patients_deaths':'Patients Deaths'}, inplace = True)
sn
```

```python
sn.to_csv('nurse.csv', index=False)
```
