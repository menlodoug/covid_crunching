```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
```

```python
apple_df = pd.read_csv("https://covid19-static.cdn-apple.com/covid19-mobility-data/2109HotfixDev19/v3/en-us/applemobilitytrends-2021-06-07.csv", low_memory=False)
apple_df = apple_df[(apple_df['geo_type'] == 'county') & (apple_df['transportation_type'] == 'driving')]
apple_df["region"]= apple_df["region"].str.replace(" County", "")
apple_df = apple_df.rename({'region' : 'County', 'sub-region' : 'State'}, axis=1)
apple_df = apple_df.drop(['geo_type', 'transportation_type','alternative_name', 'country'], axis=1)
apple_df.head()
```

```python
apple_df = (apple_df.set_index(['County', 'State'])
   .rename_axis(['Date'], axis=1)
   .stack()
   .reset_index())
apple_df
```

```python
apple_df = apple_df.rename({0: 'Index'}, axis='columns')
apple_df
```

```python
apple_df.to_csv('apple.csv', index=False)
```
