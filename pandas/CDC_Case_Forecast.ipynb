```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.0f}'.format
```

```python
case_forecast_df = pd.read_csv('2021-11-29-all-forecasted-cases-model-data.csv')
case_forecast_df.head(5)
```

```python
case_forecast_df['location_name'] = case_forecast_df['location_name'].apply(lambda x: x.split(' County,')[0])
case_forecast_df = case_forecast_df.fillna(0)
case_forecast_df = case_forecast_df[(case_forecast_df['target'] != '5 wk ahead inc case')]
case_forecast_df = case_forecast_df[(case_forecast_df['target'] != '6 wk ahead inc case')]
case_forecast_df = case_forecast_df[(case_forecast_df['target'] != '7 wk ahead inc case')]
case_forecast_df = case_forecast_df[(case_forecast_df['target'] != '8 wk ahead inc case')]
case_forecast_df.head(5)
```

```python
case_forecast_df = case_forecast_df.query("State != location_name")
case_forecast_df
```

```python
case_forecast_df.to_csv('case_forecast.csv', index=False)
```
