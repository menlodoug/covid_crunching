```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.0f}'.format
```

```python
scot_shot = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/country_data/Scotland.csv')
scot_shot.head()
```

```python
scot_vac = scot_shot[['date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'location']]
scot_vac.rename(columns={'location' : 'country'}, inplace=True)
scot_vac
```

```python
scot_vac.to_csv('scot_shot.csv', index=False)
```
