```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
```

```python
jail_df = pd.read_csv("https://media.githubusercontent.com/media/uclalawcovid19behindbars/data/master/latest-data/latest_facility_counts.csv", low_memory=False)
jail_df = jail_df[['Facility.ID', 'Jurisdiction', 'State', 'Name', 'Address', 'City', 'Zipcode', 'County', 'County.FIPS', 'Latitude', 'Longitude','Residents.Confirmed', 'Staff.Confirmed', 'Residents.Deaths', 'Staff.Deaths', 'Date']].fillna(0)
jail_df['Date'] = pd.to_datetime(jail_df['Date'])
jail_df['Address'] = jail_df['Address'].str.title()
jail_df['City'] = jail_df['City'].str.title()
jail_df['County'] = jail_df['County'].str.title()
jail_df[['Facility.ID','Zipcode', 'County.FIPS', 'Staff.Confirmed', 'Residents.Deaths', 'Staff.Deaths']] = jail_df[['Facility.ID', 'Zipcode','County.FIPS', 'Staff.Confirmed', 'Residents.Deaths', 'Staff.Deaths']].astype(int)
jail_df
```

```python
jailed = jail_df
jailed2 = jailed[(jailed['Facility.ID'] != 0)]
jailed2 = jailed2.sort_values(by=['Facility.ID'])
jailed2.to_csv('jailed2.csv', index=False)
```

```python
#jail_df.to_csv('jail.csv', index=False)
jail = jail_df[(jail_df['Facility.ID'] == 0)]
jail.rename({'Facility.ID': 'ID'}, axis=1, inplace=True)
jail
```

```python
jail.to_csv('no_id.csv', index=False)
```

```python
summed = jailed2.groupby(['Facility.ID', 'Date']).sum()
listed = summed.reset_index(drop=False)
listed.rename({'Facility.ID': 'ID'}, axis=1, inplace=True)
listed
```

```python
listed.to_csv('jailed.csv', index=False)
```
