```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
```

```python
race_df = pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-race-ethnicity.csv')
race_df
```

```python
race_df.replace(to_replace =["other", "cdph-other"], value ="Other", inplace=True)
race_df.replace(to_replace ="asian", value ="Asian", inplace=True)
race_df.replace(to_replace ="black", value ="Black", inplace=True)
race_df.replace(to_replace ="latino", value ="Latino", inplace=True)
race_df.replace(to_replace ="white", value ="White", inplace=True)
race_df.replace(to_replace ="all", value ="Total", inplace=True)
race_df
```

```python
race_df[race_df.age != '18+']
```

```python
g = race_df.groupby(['date', 'race', 'age'])['confirmed_cases_total', 'confirmed_cases_percent', 'deaths_total', 'deaths_percent', 'population_percent'].sum().reset_index()
g
```

```python
g.to_csv('races.csv', index=False)
```
