```python
totals = pd.merge(cases, deaths, on=['Province','Country','Date'])
totals
```

```python
totals['Date'] = pd.to_datetime(totals['Date'])
totals
```

```python
totals1=totals.sort_values(by = ['Country', 'Province', 'Date'], ascending = [True, True, True])
totals1
```

```python
totals1.sort_values(by=['Country', 'Province', 'Date'], inplace=True)
totals1['Case_Day'] = totals1.groupby('Country').Cases.diff().fillna(totals1.Cases)
totals1['Case_Day'].fillna(0, inplace=True)
totals1
```

```python
totals1.sort_values(by=['Country', 'Province', 'Date'], inplace=True)
totals1['Death_Day'] = totals1.groupby('Country').Deaths.diff().fillna(totals1.Deaths)
totals1['Death_Day'].fillna(0, inplace=True)
totals1
```

```python
cols = ['Case_Day', 'Death_Day']
totals1[cols] = totals1[cols].applymap(np.int64)
totals1
```

```python
continent_df = pd.read_csv('country_continent.csv')
continent_df
```

```python
global_df = pd.merge(totals1, continent_df, on='Country', how='outer')
global_df[['Cases', 'Deaths', 'Case_Day', 'Death_Day']] = global_df[['Cases', 'Deaths', 'Case_Day', 'Death_Day']].fillna(0)
global_df[['Cases', 'Deaths', 'Case_Day', 'Death_Day']] = global_df[['Cases', 'Deaths', 'Case_Day', 'Death_Day']].astype(int)
global_df
```
