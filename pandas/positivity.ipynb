```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
```

```python
positive_df = pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-positive-test-rate.csv')
positive_df.tail(5)
```

```python
bool_series = pd.notnull(positive_df["new_tests_seven_day_total"])
positivity = positive_df[bool_series]
positivity['test'] = 'COVID'
positivity[['confirmed_cases','total_tests','new_confirmed_cases', 'new_positive_tests','new_tests','new_confirmed_cases_seven_day_total','new_tests_seven_day_total']] = positivity[['confirmed_cases','total_tests','new_confirmed_cases', 'new_positive_tests','new_tests','new_confirmed_cases_seven_day_total','new_tests_seven_day_total']].astype(int)
positivity
```

```python
positivity.to_csv('positivity.csv', index=False)
```
